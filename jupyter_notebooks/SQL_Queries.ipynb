```python
# Import Libraries
import numpy as np
import pandas as pd
import sqlite3
from sqlite3 import Error
import warnings
warnings.filterwarnings('ignore')
```

```python
# Function to create a new SQLite3 Database
def create_connection(db_file):
    try:
        conn = sqlite3.connect(db_file)
    except Error as e:
        print(e)
    finally:
        conn.close()
```

```python
# Create a connection object using the connect function 
# To establish a connection with "citibikes.db"
connection = sqlite3.connect("spotify_project.db")
# Create a cursor object by calling the cursor method
cursor = connection.cursor()
```

```python
pd.read_sql_query('''SELECT * FROM spotify_songs''', connection)
```

```python
pd.read_sql_query('''SELECT * FROM spotify_numbers''', connection)
```

```python
songs_by_playlist = pd.read_sql_query('''SELECT track_name, danceability, COUNT(track_name) 
                    FROM spotify_songs GROUP BY track_name 
                    ORDER BY COUNT(track_name) DESC LIMIT 200''', connection)
```

```python
playlist_by_danceability = pd.read_sql_query('''SELECT playlist_name, danceability, 
                    COUNT(danceability) 
                    FROM spotify_songs 
                    WHERE danceability = "Dance like nobody's watching"
                    GROUP BY playlist_name 
                    ORDER BY COUNT(danceability) DESC LIMIT 200''', connection)
```

```python
playlist_rap = pd.read_sql_query('''SELECT playlist_name, track_name, danceability  
                    FROM spotify_songs 
                    WHERE playlist_name = "Rap"''', connection)
```

```python
playlist_party = pd.read_sql_query('''SELECT playlist_name, track_name, danceability  
                    FROM spotify_songs 
                    WHERE playlist_name = "Party"''', connection)
```

```python
most_danceable_year = pd.read_sql_query('''SELECT release_date,
                    COUNT(danceability)
                    FROM spotify_songs 
                    WHERE danceability = "Dance like nobody's watching"
                    GROUP BY release_date 
                    ORDER BY COUNT(danceability) DESC LIMIT 60''', connection)
```

```python
most_energetic_year = pd.read_sql_query('''SELECT release_date,
                    SUM(energy) 
                    FROM spotify_songs 
                    GROUP BY release_date 
                    ORDER BY SUM(energy) DESC LIMIT 60''', connection)
```

```python
categories_danceability = pd.read_sql_query('''SELECT danceability, COUNT(danceability) 
                    FROM spotify_songs GROUP BY danceability 
                    ORDER BY COUNT(danceability) DESC''', connection)
```

```python
songs_by_popularity = pd.read_sql_query('''SELECT track_name, artist_name, popularity, danceability,
                    SUM(popularity) 
                    FROM spotify_songs 
                    GROUP BY track_name 
                    ORDER BY SUM(energy) DESC LIMIT 200''', connection)
```
